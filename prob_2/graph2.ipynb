```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#check if data is .csv or .xlsx
data = '/home/harsh/Documents/trustlab/fresh_selenium/prob_2/dataset.csv'
if data.endswith('.csv'):
    df = pd.read_csv(data, header = None)
elif data.endswith('.xlsx'):
    df = pd.read_excel(data,header = None)
```

```python
df
```

```python
df = df
```

```python
df = df.iloc[12:]
df
```

```python
df = df.drop([ 4,5,6, 7, 8, 9, 10], axis=1)
df
```

```python
df_up = df.iloc[0:7]
df_up
```

```python
df_down = df.iloc[8:]
df_down
```

```python
df_up.reset_index(drop=True, inplace=True)
df_up
```

```python
# reset index for df_down 
df_down.reset_index(drop=True, inplace=True)
df_down
```

```python
columns = ['Age', 'Unvaccinated', '1-dose', '2-dose']
df_up.columns = columns
df_up
```

```python
# for down
columns = ['Age', 'Unvaccinated', '1-dose', '2-dose']
df_down.columns = columns
df_down
```

```python
# for age below 45 years
df_up["Unvaccinated"] = df_up.Unvaccinated.astype(float)
df_up['1-dose'] = df_up['1-dose'].astype(float)
df_up['2-dose'] = df_up['2-dose'].astype(float)
df_up.dtypes

# for age above 45 years
df_down["Unvaccinated"] = df_down.Unvaccinated.astype(float)
df_down['1-dose'] = df_down['1-dose'].astype(float)
df_down['2-dose'] = df_down['2-dose'].astype(float)
df_down.dtypes
```

```python
# Reverse the age_groups list to display in descending order
age_groups = df_up['Age']
unvaccinated = df_up['Unvaccinated']
dose_1 = df_up['1-dose']
dose_2 = df_up['2-dose']
age_groups = age_groups[::-1]

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting horizontal bars with different colors
bar_width = 0.25
bar_positions_unvaccinated = [pos - bar_width for pos in range(len(age_groups))]
bar_positions_dose1 = [pos for pos in range(len(age_groups))]
bar_positions_dose2 = [pos + bar_width for pos in range(len(age_groups))]

# Replace 0 with '10-14yr', 1 with '15-19yr', and so on on the y-axis
ax.set_yticks(range(len(age_groups)))
ax.set_yticklabels(age_groups)

bar1 = ax.barh(bar_positions_unvaccinated, unvaccinated[::-1], height=bar_width, label='Unvaccinated', color='blue')
bar2 = ax.barh(bar_positions_dose1, dose_1[::-1], height=bar_width, label='1-dose', color='orange')
bar3 = ax.barh(bar_positions_dose2, dose_2[::-1], height=bar_width, label='2-dose', color='red')

# Customize the plot
ax.set_xlabel('Death rate per 100,000 person-year')
ax.set_ylabel('Age Groups')
ax.set_title('Age Groups under 45 years')
ax.legend()

# Show the plot
plt.show()
```
