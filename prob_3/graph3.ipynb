```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', sep=';')
df
```

```python
df['value'] = (df['value'] / 1000)
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bar graph
bars = ax.bar(df['year'], df['value'], color='blue')

# Customize the plot
ax.set_xlabel('Year')
ax.set_ylabel('Values')
ax.set_title('Bar Graph for Values vs Year')

# Show the plot
plt.show()
```
